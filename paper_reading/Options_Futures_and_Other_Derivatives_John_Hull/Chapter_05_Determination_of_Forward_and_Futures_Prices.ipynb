# Chapter 5: Determination of Forward and Futures Prices

> **Color Convention**  
> 🟡 **Yellow** = Definition / Theorem / Property  
> 🟢 **Green** = Comment / Application / Insight

---

## Overview

This chapter derives the no-arbitrage forward and futures prices for various underlying assets: non-income-paying assets, income-paying assets, stock indices, currencies, and commodities. It also discusses the relationship between futures prices and expected future spot prices.

---

## 5.1 Investment Assets vs. Consumption Assets

🟡 **Definition:**

> An **investment asset** is held by at least some investors purely for investment (e.g., stocks, bonds, gold, silver). The no-arbitrage forward price can be derived precisely.
>
> A **consumption asset** is held primarily for consumption, not investment (e.g., oil, corn, copper). Storage allows holding the asset, but the asset cannot always be shorted freely → only an upper bound on the forward price can be established.

---

## 5.2 Short Selling

🟡 **Property:**

> **Short selling** involves borrowing and selling an asset you don't own, then buying it back later to return it. The short seller must pay any dividends or income that the asset generates to the lender.
>
> The short seller must maintain a **margin account** to guarantee the position — similar to the margin account for futures.

---

## 5.3 Assumptions and Notation

- $S_0$ = spot price today
- $F_0$ = forward/futures price today
- $T$ = time to delivery (years)
- $r$ = continuously compounded risk-free rate
- $q$ = continuously compounded dividend/income yield
- $u$ = storage cost (as proportion of price per year)
- $y$ = convenience yield

---

## 5.4 Forward Price for an Investment Asset

### No Income

$$\boxed{F_0 = S_0 e^{rT}}$$

**Proof by arbitrage:**
- If $F_0 > S_0 e^{rT}$: borrow $S_0$, buy asset, short forward → lock in profit at $T$
- If $F_0 < S_0 e^{rT}$: short asset, invest proceeds, long forward → profit at $T$

### Known Cash Income (Present Value $I$)

$$F_0 = (S_0 - I)e^{rT}$$

### Known Yield $q$ (Continuous)

$$F_0 = S_0 e^{(r-q)T}$$

---

## 5.5 Stock Index Futures

🟡 **Property:**

> For a stock index paying continuous dividend yield $q$:
>
> $$F_0 = S_0 e^{(r-q)T}$$
>
> **Index arbitrage**: If $F_0 > S_0 e^{(r-q)T}$, buy the index stocks and short the futures. If $F_0 < S_0 e^{(r-q)T}$, short the stocks and go long futures. Index arbitrage is typically implemented via **program trading** (computer-generated orders).

---

## 5.6 Currency Futures (Interest Rate Parity)

For a foreign currency with foreign risk-free rate $r_f$:

$$F_0 = S_0 e^{(r-r_f)T}$$

This is the **covered interest parity** (or interest rate parity) condition. A currency is like an investment asset paying a continuous yield equal to the foreign risk-free rate.

---

## 5.7 Futures on Commodities

### Gold and Silver

🟡 **Property:**

> Gold and silver are investment commodities. Gold owners (e.g., central banks) can lend gold at the **gold lease rate** $\ell$:
>
> $$F_0 = S_0 e^{(r + u - \ell)T}$$
>
> where $u$ = storage cost as a continuous rate. If $\ell = 0$: $F_0 = S_0 e^{(r+u)T}$.

### Consumption Commodities

🟡 **Property (Convenience Yield):**

> For consumption commodities (oil, gas, corn), the futures price satisfies:
>
> $$F_0 = S_0 e^{(r+u-y)T}$$
>
> where $y$ = **convenience yield** (the benefit of holding the physical commodity — e.g., ability to profit from temporary local shortages, to keep a production process running). A futures contract does not provide this benefit.
>
> If $y > r + u$: **backwardation** (futures below spot).  
> If $y < r + u$: **contango** (futures above spot).

### Cost of Carry

🟡 **Definition:**

> The **cost of carry** $c$ = storage cost + interest cost − income earned:
>
> | Asset | Cost of Carry $c$ |
> |-------|------------------|
> | Non-dividend stock | $r$ |
> | Stock with yield $q$ | $r - q$ |
> | Currency | $r - r_f$ |
> | Gold | $r + u$ |
> | Commodity | $r + u - y$ |
>
> $$F_0 = S_0 e^{cT}$$

---

## 5.8 Futures Prices and Expected Future Spot Prices

🟡 **Definition:**

> - **Normal backwardation**: $F_0 <$ expected future spot price $E[S_T]$
> - **Contango**: $F_0 >$ $E[S_T]$
>
> (Note: these terms are also sometimes used to describe whether the futures price is below or above the *current* spot price — usage varies.)

Under the **risk-premium** theory:
- If the underlying asset is positively correlated with the market (systematic risk), futures speculators going long require compensation → $F_0 < E[S_T]$ → **normal backwardation**
- If negatively correlated with the market, $F_0 > E[S_T]$ → **contango**

For risk-neutral investors ($\lambda = 0$): $F_0 = E[S_T]$.

---

## Summary

| Asset | Forward Price Formula |
|-------|----------------------|
| No income | $F_0 = S_0 e^{rT}$ |
| Known income (PV = $I$) | $F_0 = (S_0-I)e^{rT}$ |
| Continuous yield $q$ | $F_0 = S_0 e^{(r-q)T}$ |
| Currency | $F_0 = S_0 e^{(r-r_f)T}$ |
| Commodity | $F_0 = S_0 e^{(r+u-y)T}$ |
