# Chapter 2 Financial Market Equilibrium

## Introduction

- The chapter states the relationship between information, prices, individuals’ decisions and market equilibrium by emphasizing that market prices both coordinate actions and transmit the knowledge needed for agents to take the right action.
- Market price plays two roles: it is an allocation mechanism that coordinates consumption and production decisions and an information signal on the future value of the asset being exchanged.

### Allocating scarce resources

- In Figure 2.1, the aggregate supply curve $S$ and aggregate demand curve $D$ determine the Walrasian equilibrium price $p^*$ and quantity $q^*$, where consumers’ and enterprises’ decisions are coordinated, information is exogenous and homogeneous, and agents do not use the equilibrium price to revise their decisions.
    - A Walrasian equilibrium is a market state where prices adjust so aggregate supply equals aggregate demand, and every agent optimizes their own choice given those prices.

### Transmitting information

- In the opening-auction example, investor $i$ receives information $y_i$, submits a buy order $x(y_i)$, observes a lower-than-expected equilibrium price $p^*(y_{-i})$ reflecting other agents’ information $y_{-i}$, and revises his demand to $x(y_i,p^*(y_{-i}))$, so **prices become a vehicle of information when information available to traders is not uniform**.
- Rational expectations equilibrium is an equilibrium in which, if the price is a vehicle of information, **the equilibrium price is such that once agents have observed it they do not wish to trade again**, or equivalently they no longer wish to alter their investment decisions.
- The rational expectations definition may lead to conclusions different from the Walrasian paradigm because prices can affect perceived asset quality, demand and supply, and the existence or informational content of equilibrium.
    - **The market demand function may not be decreasing** because a price reduction may signal a lower value of the security and cause a decline rather than an increase in trading demand.
    - **There may be interdependence between the demand and supply functions** because an increase in supply can reduce the security price and, since lower prices correspond on average to states where the security value is lower, can worsen uninformed agents’ valuation and reduce their demand.
    - **There may be no equilibrium** because if all available information can be obtained from prices, costly information acquisition becomes unattractive, but if fewer agents acquire information prices lose informational content and again create incentives to acquire information.
        - A **not fully revealing equilibrium** can exist when noise traders or endowment shocks prevent uninformed agents from isolating insiders’ signals simply by looking at price, although this equilibrium is not robust when agents are nearly risk-neutral or when market noise is too small.
        - A **fully revealing equilibrium** arises when the equilibrium price reveals all available information, in which case agents can infer the insiders’ signal from price and the incentive to acquire costly information disappears.
- **Imperfect competition**, introduced by Kyle, means that agents factor in the effects of their own investment strategies on price and therefore submit less aggressive orders than under competitive behavior.
    - First, under imperfect competition prices cannot reveal more than half the information available to insiders and the information revealed is always less than under competitive behavior.
    - Second, with imperfect competition an equilibrium always exists even in the presence of informed risk-neutral agents, which justifies the broad use of models allowing for strategic interaction.

## Financial market equilibrium with symmetric information

### A single risky asset

- The representative agent chooses risky-asset demand $X$ to maximize expected utility of end-of-period wealth according to $$\max_X E[U(\tilde w)],\qquad \tilde w=(I+X)\tilde F+(I^f-Xp)(1+r_f),$$ where $I$ is the initial endowment of the risky asset, $I^f$ is the initial endowment of the risk-free asset, $\tilde F$ is the normally distributed risky-asset payoff, $p$ is the risky-asset price, $r_f$ is the risk-free rate, and the equation says wealth equals the final payoff from total risky holdings plus the payoff from the risk-free position.
- Starting from the first-order condition $E[u'(\tilde w)(\tilde F-p(1+r_f))]=0$, using covariance decomposition and Stein’s lemma gives $$E[u'(\tilde w)]E[\tilde F-p(1+r_f)]+E[u''(\tilde w)](I+X)\operatorname{Var}(\tilde F)=0,$$ so the equilibrium price is $$p=\frac{1}{1+r_f}\left\{E(\tilde F)+\frac{E[u''(\tilde w)]}{E[u'(\tilde w)]}(I+X)\operatorname{Var}(\tilde F)\right\},$$ where $u'$ and $u''$ are marginal utility and its derivative, and the equation prices the risky asset by discounting expected payoff plus the covariance-based risk adjustment required by risk aversion and aggregate risky holdings.
- With CARA utility $u(\tilde w)=-\exp(-A\tilde w)$, the pricing equation becomes $$p=\frac{1}{1+r_f}\left\{E(\tilde F)-A(I+X)\operatorname{Var}(\tilde F)\right\},$$ where $A$ is the coefficient of absolute risk aversion, so the price of the asset is its discounted expected cash flow less a discount for risk that depends on payoff variance, risky-asset endowment, and risk aversion.

### Multiple risky assets and the CAPM

- For asset $i$, the same first-order-condition, covariance, and Stein’s-lemma argument gives $$p_i=\frac{1}{1+r_f}\left\{E(\tilde F_i)+\frac{E[u''(\tilde w)]}{E[u'(\tilde w)]}\operatorname{Cov}(\tilde w,\tilde F_i)\right\},$$ where $p_i$ is asset $i$’s price, $\tilde F_i$ is its cash flow, and $\operatorname{Cov}(\tilde w,\tilde F_i)$ measures how its payoff covaries with total wealth, so price equals discounted expected cash flow plus a risk-premium term based on covariance with wealth.
- The expected return equation is $$E(\tilde r_i)=r_f+\left\{-\frac{E[u''(\tilde w)]}{E[u'(\tilde w)]}\right\}\operatorname{Cov}(\tilde w,\tilde r_i),$$ where $\tilde r_i=(\tilde F_i-p_i)/p_i$, and after defining the market portfolio return $\tilde r_m$, using $$-\frac{E[u''(\tilde W)]}{E[u'(\tilde W)]}=\frac{E(\tilde r_m)-r_f}{\operatorname{Var}(\tilde r_m)p_m}$$ yields $$E(\tilde r_i)-r_f=\frac{\operatorname{Cov}(\tilde r_m,\tilde r_i)}{\operatorname{Var}(\tilde r_m)}[E(\tilde r_m)-r_f]=\beta_i[E(\tilde r_m)-r_f],$$ which is consistent with CAPM because each asset’s expected excess return is proportional to its beta with the market portfolio.

## Financial market equilibrium with asymmetric information

- With CARA utility $E[U(\tilde\pi)]=E[-e^{-A\tilde\pi}]$ and profit $\tilde\pi=X(\tilde F-p)$, normality implies that maximizing expected utility is equivalent to maximizing $E(\tilde\pi)-\frac{A}{2}\operatorname{Var}(\tilde\pi)$, so an uninformed trader’s demand is $$X=\frac{E(\tilde F)-p}{A\operatorname{Var}(\tilde F)}$$ and an informed trader’s demand is $$X_I=\frac{E(\tilde F\mid S)-p}{A\operatorname{Var}(\tilde F\mid S)},$$ where $S$ is the private signal and the strategy says traders buy more when expected payoff exceeds price and less when risk aversion or conditional variance is high.

### Naive expectations

- Plugging the informed demand $X_I$, uninformed demand $X$, and liquidity order $\tilde x$ into the market-clearing condition $NX_I+MX+Z\tilde x=0$ gives $$p=\mu_1E(\tilde F\mid S)+(1-\mu_1)E(\tilde F)+\mu_2\tilde x,$$ where $$\mu_1=\frac{N\operatorname{Var}(\tilde F)}{N\operatorname{Var}(\tilde F)+M\operatorname{Var}(\tilde F\mid S)},\qquad \mu_2=\frac{AZ\operatorname{Var}(\tilde F)\operatorname{Var}(\tilde F\mid S)}{N\operatorname{Var}(\tilde F)+M\operatorname{Var}(\tilde F\mid S)},$$ and the equation says that under naïve expectations the equilibrium price is a weighted average of informed and uninformed expected cash flows plus a noise-trader risk-premium term because uninformed traders use only their prior rather than the information contained in prices.

### Rational expectations

- A linear rational expectations equilibrium requires that uninformed agents conjecture a linear price rule $\tilde p=\alpha_1\tilde S+\alpha_2\tilde x$, use this conjecture to compute $E(\tilde F\mid p)$ and $\operatorname{Var}(\tilde F\mid p)$ and hence their demand, and reach equilibrium when the market-clearing price $NX_I+MX+Z\tilde x=0$ is itself equal to the conjectured linear combination of $\tilde S$ and $\tilde x$.
- With deterministic non-random supply $x$, uninformed traders can infer the signal perfectly from the naïve price rule, so their demand becomes $X=X_I=[E(\tilde F\mid S)-p]/[A\operatorname{Var}(\tilde F\mid S)]$ and the fully revealing rational expectations price is $$p=E(\tilde F\mid S)-A\operatorname{Var}(\tilde F\mid S)\frac{Zx}{N+M},$$ where $N$ and $M$ are the numbers of informed and uninformed traders and $Zx$ is aggregate liquidity demand, so price equals the signal-conditional expected payoff less a discount for aggregate risk aversion.
- In the noisy rational expectations model, the market-clearing condition is $$NX_I+MX+Z\tilde x=0,$$ where $X_I=(\tilde S-p)/(A\sigma_\varepsilon^2)$ for insiders, $X$ is the uninformed demand based on information extracted from price, and $Z\tilde x$ is aggregate noise-trader demand.
- If uninformed traders conjecture that other uninformed traders submit $X=-Hp$, extract the residual signal $\tilde\xi=\tilde S+(A\sigma_\varepsilon^2Z/N)\tilde x$, and impose that conjectures coincide with realizations, then $$H^*=\frac{1-\operatorname{Cov}(\tilde F,\tilde\xi)/\operatorname{Var}(\tilde\xi)}{A\operatorname{Var}(\tilde F\mid\tilde\xi)+[\operatorname{Cov}(\tilde F,\tilde\xi)/\operatorname{Var}(\tilde\xi)](MA\sigma_\varepsilon^2/N)}$$ and the noisy rational expectations equilibrium price is $$p^*=\left\{\frac{N}{A\sigma_\varepsilon^2}+MH^*\right\}^{-1}\left\{\frac{N}{A\sigma_\varepsilon^2}\tilde S+Z\tilde x\right\},$$ where $\sigma_\varepsilon^2$ is the residual variance in $\tilde F=\tilde S+	ilde\varepsilon$, and the equation says price aggregates insiders’ signal and noise trade while expectations coincide with the true realization.
- Once $p^*$ is determined, liquidity, volatility, and informational efficiency are measured by $$L=\left|\frac{dp}{dx}\right|^{-1}=\frac{N}{A\sigma_\varepsilon^2}+MH^*,$$ $$\operatorname{Var}(p^*)=L^{-2}\left\{\frac{N^2}{A^2\sigma_\varepsilon^4}\sigma_S^2+Z^2\sigma_x^2\right\},$$ and $$IE=\operatorname{Var}(\tilde F\mid\tilde\xi)^{-1}=\left(\sigma_S^2+\sigma_\varepsilon^2-\frac{\sigma_S^4}{\sigma_S^2+(A^2\sigma_\varepsilon^4Z^2/N^2)\sigma_x^2}\right)^{-1},$$ where $L$ is inverse price impact or market depth, $\operatorname{Var}(p^*)$ is price volatility, and $IE$ is inverse residual variance after conditioning on the signal extracted from equilibrium price.

## Financial market equilibrium with asymmetric information and strategic agents

- The model further assumes that before trading informed traders receive an endowment shock $\tilde I\sim N(0,\sigma_I^2)$, which is unknown to the uninformed and lets insiders trade both to speculate on private information and to hedge their endowment of risky assets.
- With strategic behavior, each informed trader submits $$X_I^S=\frac{\tilde S-p^S-A\sigma_\varepsilon^2\tilde I}{A\sigma_\varepsilon^2+\lambda_I}=D(\tilde S-p^S)-G\tilde I,$$ where $D=1/(A\sigma_\varepsilon^2+\lambda_I)$, $G=A\sigma_\varepsilon^2/(A\sigma_\varepsilon^2+\lambda_I)$, and $\lambda_I=\partial p^S/\partial X_I^S$ is the price impact of the informed trader’s order, and solving the strategic model gives $$p^S=(L^SA)^{-1}[ND\tilde S-NG\tilde I+Z\tilde x]$$ and $$L^SA=\left|\frac{dp}{dx}\right|^{-1}=ND+MH^{S*},$$ so the equilibrium price reflects the private signal, the informed traders’ endowment shock, and noise-trader orders while price impact is the inverse of strategic market liquidity.
- Traders’ strategic behaviour diminishes their willingness to supply liquidity because they internalize the price impact of their own demand, scale back their orders, and are less aggressive than competitive traders in accommodating a noise trader’s order.
- Informed traders should be considered as the best liquidity suppliers because the numerical simulations show that liquidity falls much more when informed traders rather than uninformed traders behave strategically, reflecting the fact that informed traders face the lowest adverse selection costs.

## Appendix

- Stein's lemma says that, for a normally distributed variable, the covariance between a function of that variable and the variable itself can be rewritten using the derivative of that function. In this chapter, it helps turn a hard term like $\operatorname{Cov}(u'(\tilde w), \tilde F)$ into something involving $E[u''(\tilde w)]\operatorname{Cov}(\tilde w, \tilde F)$, so economically it connects marginal utility, risk, and asset payoff covariance in a clean pricing equation.
- CAPM stands for Capital Asset Pricing Model. It says an asset's expected excess return depends on how much that asset moves with the overall market, according to $$E(r_i)-r_f=\beta_i[E(r_m)-r_f],$$ so investors are rewarded only for systematic market risk, not for risk that can be diversified away.
- CARA utility means constant absolute risk aversion, usually written as $u(w)=-e^{-Aw}$, where $A>0$ measures how risk-averse the trader is. In this chapter, CARA utility makes risky-asset demand linear in the gap between expected payoff and price, while higher $A$ reduces demand because the trader dislikes payoff uncertainty more strongly.
